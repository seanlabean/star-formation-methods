```python
def extract_seba_data(input_model_file, input_star_file):
    dat_raw = open(input_model_file,'r').read().split("\n")
    evolve_model = []
    for entry in dat_raw:
        entry = entry.split("\n")
        entry = " ".join(entry)   # Unpacks list into a single string and adds " " to end of string.

        entry = entry.split(" ")  # Splits the string into indiv elems based on " " separating entries
                                  # And now we're good to go!
        evolve_model.append(entry)


    dat_raw = open(input_star_file,'r').read().split("\n")
    evolve_star = []
    for entry in dat_raw:
        entry = entry.split("\n")
        entry = " ".join(entry)   # Unpacks list into a single string and adds " " to end of string.

        entry = entry.split(" ")  # Splits the string into indiv elems based on " " separating entries
                                  # And now we're good to go!
        evolve_star.append(entry)


    e_modl_time = []
    e_modl_mass = []
    e_modl_radi = []
    e_modl_lumi = []
    e_modl_temp = []

    e_star_time = []
    e_star_mass = []
    e_star_radi = []
    e_star_lumi = []
    e_star_temp = []

    for line in evolve_model:
        try:
            e_modl_time.append(float(line[0]))
            e_modl_mass.append(float(line[1]))
            e_modl_radi.append(float(line[2]))
            e_modl_lumi.append(float(line[3]))
            e_modl_temp.append(float(line[4]))
        except:
            pass
    for line in evolve_star:
        try:
            e_star_time.append(float(line[0]))
            e_star_mass.append(float(line[1]))
            e_star_radi.append(float(line[2]))
            e_star_lumi.append(float(line[3]))
            e_star_temp.append(float(line[4]))
        except:
            pass
    return e_modl_time, e_modl_mass, e_modl_radi, e_modl_lumi, e_modl_temp, e_star_time, e_star_mass, e_star_radi, e_star_lumi, e_star_temp
```
